```python
import pandas as pd
import numpy as np 
from scipy.ndimage import gaussian_filter1d
import matplotlib.pyplot as plt


obj = 'tofu' #'tofu' #'watermelon'
iteration = 0
def read_tactile(fname):
    df = pd.read_csv( fname, names=['isPos', 'taxel', 'adc', 't'], sep=' ')

    df.t -= df.t.iloc[0]

    df.drop(df.tail(1).index,inplace=True)
    df.drop(df.head(1).index,inplace=True)
    #print(df.taxel)
    df = df.loc[ (df.taxel >= 1) & (df.taxel <= 80) ]
    #print(df.shape)
    df = df.reset_index(drop=True)

    return df



left_adc = np.array([36, 38, 40, 32, 34, 26, 28, 30, 22, 24,5,3,1, 9,7, 15,13, 11, 19, 17])
# left_fa = np.setdiff1d(left_taxels, left_adc)
right_adc = np.array([56, 58, 60, 52, 54, 46, 48, 50, 42, 44, 65, 63,61, 69, 67, 75, 73, 71, 79, 77])
# right_fa = np.setdiff1d(right_taxels, right_adc)


df = read_tactile(f'{obj}_zero_{iteration}.tact')

#df_left = df[df.taxel.isin(left_adc)]
#df_right = df[df.taxel.isin(right_adc)]


fig, ax = plt.subplots(ncols=2, figsize=(16,8))
for taxel in left_adc:
    sample = df[df.taxel == taxel]

    if sample.adc.max() >= 300:
        continue

    if len(sample) != 0:
        ax[0].plot(sample.t, sample.adc, label=str(taxel))
        
        # play with a sample adc
        adc = sample.adc.values
        t   = sample.t.values
#         print(t)
        ax[0].plot(t,adc)
        adc = gaussian_filter1d(adc, sigma = 1.5)
        plt.plot(t, adc)
        spike = get_spike(adc, t)

        ax[0].scatter(t[spike], adc[spike])

        ax[0].legend()
for taxel in right_adc:
    sample = df[df.taxel == taxel]

    if sample.adc.max() >= 300:
        continue

    if len(sample) != 0:
        ax[1].plot(sample.t, sample.adc, label=str(taxel))
        
        # play with a sample adc
        adc = sample.adc.values
        t   = sample.t.values
        ax[1].plot(t,adc)
        spike = get_spike(adc, t)

        ax[1].scatter(t[spike], adc[spike])
        ax[1].legend()
```

```python
def get_spike(adc, t):
    def moving_average(x, w):
        return np.convolve(x, np.ones(w), 'valid') / w
    k = 5
    move_average = moving_average(adc, k)
    adc = gaussian_filter1d(adc, sigma = 0.9)

    for i in range(k,len(adc)):
#         print((adc[i]-moving_average)/moving_average)
        if abs(adc[i]-move_average[i-k])/move_average[i-k] > 0.006:
#             print("here")
            if t[i] < 2:
                continue
            return i
            

    
    
    
    return 0
```
